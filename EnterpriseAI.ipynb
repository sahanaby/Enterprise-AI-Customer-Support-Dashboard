```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
data = {
    "Ticket_ID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],

    "Customer_Name": [
        "Rahul","Anita","Kiran","Meena","Ravi",
        "Sita","Arjun","Neha","Vijay","Pooja"
    ],

    "Issue_Type": [
        "Billing","Technical","Account","Technical","Billing",
        "Account","Technical","Billing","Account","Technical"
    ],

    "Priority": [
        "High","Medium","Low","High","Medium",
        "Low","High","Medium","Low","High"
    ],

    "Status": [
        "Open","Closed","Pending","Open","Closed",
        "Pending","Open","Closed","Pending","Open"
    ],

    "Response_Time_Hours": [
        2,5,8,3,6,7,2,5,9,1
    ]
}

df = pd.DataFrame(data)

print("Ticket Dataset:")
df
```

```python
total_tickets = len(df)

open_tickets = df[df["Status"] == "Open"].shape[0]

closed_tickets = df[df["Status"] == "Closed"].shape[0]

avg_response = df["Response_Time_Hours"].mean()

print("Total Tickets:", total_tickets)
print("Open Tickets:", open_tickets)
print("Closed Tickets:", closed_tickets)
print("Average Response Time:", round(avg_response, 2))
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Issue_Type",
    data=df
)

plt.title("Tickets by Issue Type")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Priority",
    data=df
)

plt.title("Priority Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Status",
    data=df
)

plt.title("Ticket Status Dashboard")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Issue_Type",
    y="Response_Time_Hours",
    data=df
)

plt.title("Response Time by Issue Type")

plt.show()
```

```python
fig = px.bar(
    df,
    x="Issue_Type",
    y="Response_Time_Hours",
    color="Priority",
    title="Customer Support Performance Dashboard"
)

fig.show()
```

```python
ticket_id = int(input("Enter Ticket ID: "))

result = df[
    df["Ticket_ID"] == ticket_id
]

if result.empty:
    print("Ticket not found")

else:
    print("Ticket Found:")
    print(result)
```

```python
name = input("Enter Customer Name: ")

result = df[
    df["Customer_Name"]
    .str.lower()
    == name.lower()
]

if result.empty:
    print("Customer not found")

else:
    print(result)
```

```python
status = input("Enter Status (Open/Closed/Pending): ")

filtered = df[
    df["Status"]
    .str.lower()
    == status.lower()
]

print(filtered)
```

```python
issue = input("Enter Issue Type (Billing/Technical/Account): ")

filtered = df[
    df["Issue_Type"]
    .str.lower()
    == issue.lower()
]

print(filtered)
```
